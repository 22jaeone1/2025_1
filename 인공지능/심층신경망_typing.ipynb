## 2개의 층

```python
from tensorflow import keras

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
dense1 = keras.layers.Dense(100, activation='sigmoid', input_shape=(784,))
dense2 = keras.layers.Dense(10, activation='softmax')
```

## 심층 신경망 만들기

```python
model = keras.Sequential([dense1, dense2])
```

```python
model.summary()
```

## 층을 추가하는 다른 방법

```python
model = keras.Sequential([
    keras.layers.Dense(100, activation='sigmoid', input_shape=(784,), name='hidden'),
    keras.layers.Dense(10, activation='softmax', name='output')
], name='패션 MNIST 모델')
```

```python
model.summary()
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(100, activation='sigmoid', input_shape=(784,)))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

## 모델 훈련

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target) # 검증 세트에서도 성능 확인
```

## 렐루 활성화 함수와 Flatten 층

## 모델 훈련 - reshape() 메서드 없음

```python
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)  # 검증 세트에서도 성능 확인
```

## 옵티마이저

```python
#옵티마이저 'sgd' 사용하면 번거로움을 필할 수 있음
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# 위의 코드와 동일한 코드임
sgd = keras.optimizers.SGD()
model.compile(optimizer=sgd, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# SGD의 학습률 기본값이 0.01 일때 이를 변경하고 할 때 매개변수에 지정하여 사용
sgd = keras.optimizers.SGD(learning_rate=0.1)
```

```python
# momentum 기본값은 0, 큰 값을 지정시 그레이디언트를 가속도처럼 사용하는 모멘텀 최적화 사용, 보통 0.9이상을 지정
# nesterov 매개변수를 기본값 False에서 True로 바꾸면 네스테로프 모멘텀 최적화(네스테로프 가속 경사)를 사용
# 대부분의 경우 네스테로프 모멘텀 최적화가 기본 확률적 경사하강법보다 더 나은 성능을 제공
sgd = keras.optimizers.SGD(momentum=0.9, nesterov=True)
```

```python
# 모델이 최적점에 가까이 갈수록 학습률을 낮출 수 있고 이렇게 하면 안정적으로 최적점에 수렴할 가능성이 높음
# 이런학습률을 적응적 학습률이라고 함, 이런 방식은 학습률 매개변수를 튜닝하는 수고를 덜 수 있음
# 대표적으로 'Adagrad'와 'RMSprop' 가 있음
adagrad = keras.optimizers.Adagrad()
model.compile(optimizer=adagrad, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
rmsprop = keras.optimizers.RMSprop()
model.compile(optimizer=rmsprop, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(28, 28)))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
# 모멘텀 최적화와 RMSprop의 장점을 접목한 것이 Adam
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```
