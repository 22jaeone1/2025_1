```python
eList = []
while True:
    c = input('[Menu]i(nput),d(elte),r(eplace),p(rint),q(uit):')

    if c == 'i':
        line = int(input('Line:'))
        content = input('Content:')
        eList.insert(line, content)
    elif c== 'p':
        for i in range(len(eList)):
            print(str(i), ':', eList[i])
    elif c == 'q':
        break
```
