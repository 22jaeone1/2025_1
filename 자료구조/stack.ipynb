```python
class Stack:
    def __init__(self):
        self.top = []

    def isEmpty(self):
        return len(self.top) == 0

    def push(self, item):
        self.top.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.top.pop(-1)

    def peek(self):
        if not self.isEmpty():
            return self.top[-1]

    def __str__(self):
        return str(self.top)
```

```python
def precedence(op):
    if op == '(':
        return 0
    if op in '+-':
        return 1
    if op in '*/':
        return 2
def infix2postfix(statement):
    output = []
    s = Stack()
    for ch in statement:
        if ch == '(':
            s.push(ch)
        elif ch == ')':
            while not s.isEmpty():
                op = s.pop()
                if op == '(':
                    break
                output.append(op)
        elif ch in '+-*/':
            while not s.isEmpty():
                op = s.peek()
                if precedence(op) >= precedence(ch):
                    output.append(s.pop())
                else:
                    break
            s.push(ch)
        else: # 피연산자
            output.append(ch)
    return output

def eval(statement):
    s = Stack()
    for ch in statement:
        if ch in '+-/*':
            v2 = s.pop()
            v1 = s.pop()
            if ch == '+':
                s.push(v1 + v2)
            elif ch == '-':
                s.push(v1 - v2)
            elif ch == '*':
                s.push(v1 * v2)
            else: # ch == '/':
                s.push(v1 / v2)
        else:
            s.push(int(ch))
    return s.pop()

postfix = infix2postfix('7-4/3+5*(7/2-3)')
print(eval(postfix))
```

```python
mapMaze = [list('111111'),
           list('e00001'),
           list('101011'),
           list('11100x'),
           list('111011'),
           list('111111')]
MazeSize = 6
def isValidPos(x,y):
    if 0 <= x < MazeSize and 0 <= y < MazeSize:
        if mapMaze[x][y] in ('0','x'):
            return True
        else:
            return False
    else:
        return False

def DFS():
    s = Stack()
    s.push((1,0))

    while not s.isEmpty():
        here = s.pop()
        print(here, end = '->')
        x, y = here
        if mapMaze[x][y] == 'x':
            return True
        mapMaze[x][y] = '.'
        if isValidPos(x-1, y): # 상
            s.push((x-1,y))
        if isValidPos(x+1, y): # 하
            s.push((x+1,y))
        if isValidPos(x, y-1): # 좌
            s.push((x,y-1))
        if isValidPos(x, y+1): # 우
            s.push((x,y+1)) # 순서 : 우 -> 좌 -> 하 -> 상
        print(s)
    return False

DFS()
```

괄호 검사

```python
class Stack :
    def __init__(self): # 생성자 함수
        self.top = []

    def push(self, item):
        self.top.append(item)

    def pop(self):
        if not self.isEmpty():
            return self.top.pop()

    def peek(self):
        return self.top[-1]

    def isEmpty(self):
       return len(self.top) == 0

    def size(self):
        return len(self.top)

    def clear(self):
        self.top = []

    def __str__(self):
        return str(self.top)

def checkBrackets(s):
    stack = Stack()
    for ch in s:
        if ch in '[{(':
            stack.push(ch)
        elif ch in ']})':
            if stack.isEmpty():
                return False
            leftB = stack.pop()
            if leftB == '[' and ch != ']' :
                return False
            if leftB == '{' and ch != '}' :
                return False
            if leftB == '(' and ch != ')' :
                return False
    return stack.isEmpty()

print(checkBrackets("if ((A[i+1]) == (i+1))"))
```
